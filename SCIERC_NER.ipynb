```python
from __future__ import print_function

import os
import re
import sys
from collections import namedtuple
from io import StringIO
from os import path
from tqdm import tqdm
import glob
from brat.tools.anntoconll import process_file
import pandas as pd
```

```python

path = "raw_data/"
directory = os.fsencode(path) #Set directory with both .wav and .txt files here

error_count = 0
sucsess_count = 0

for file in tqdm(os.listdir(directory)):
    filename = os.fsdecode(file)
    
    if filename.endswith(".txt"):
      try:
          print(path+filename)
          process_file(path+filename)
          sucsess_count +=1
      except:
          error_count += 1
    
```

```python
path = "./conll/raw_data/"
directory = os.fsencode(path) #Set directory with both .wav and .txt files here
all_data_df = pd.DataFrame()

all_lines = list()

for file in os.listdir(directory):
    filename = os.fsdecode(file)    
    if filename.endswith(".txt"):
        with open(path+filename) as f:
            content = f.readlines()
            all_lines.extend(content)
```

```python
!touch finfile.txt
```

```python
with open('finfile.txt', 'a') as the_file:
    for line in all_lines:
        if line != '\n':
            line = line.split('\t')
            line = [line[-1][:-1], line[0]]
            #print(new_line)
            the_file.write((line[0]+' '+line[1]+'\n'))
        else:
            the_file.write(line)
```

```python
from simpletransformers.ner import NERModel, NERArgs
```

```python

model_args = NERArgs()
model_args.labels_list = ["O", "B-Generic", "B-Task", "B-Method", "B-OtherScientificTerm", "B-Metric", "B-Material", 
                        "I-Generic", "I-Task", "I-Method", "I-OtherScientificTerm", "I-Metric", "I-Material"]
model_args.num_train_epochs = 7
model_args.overwrite_output_dir = True
# model_args.evaluate_during_training_verbose = True
# model_args.evaluate_during_training = True

model = NERModel(
  "roberta", #electra
  "roberta-base",
  args=model_args
)

model.train_model("/content/drive/MyDrive/Diplom/finfile.txt", eval_data='/content/drive/MyDrive/Diplom/finfile.txt')
result, model_outputs, wrong_preds = model.eval_model("/content/drive/MyDrive/Diplom/finfile.txt")
print(result)
```

```python

model_args = NERArgs()
model_args.labels_list = ["O", "B-Generic", "B-Task", "B-Method", "B-OtherScientificTerm", "B-Metric", "B-Material", 
                        "I-Generic", "I-Task", "I-Method", "I-OtherScientificTerm", "I-Metric", "I-Material"]
model_args.num_train_epochs = 7
model_args.overwrite_output_dir = True
# model_args.evaluate_during_training_verbose = True
# model_args.evaluate_during_training = True

model = NERModel(
  "roberta", #electra
  "roberta-base",
  args=model_args
)

model.train_model("finfile.txt", eval_data='finfile.txt')
result, model_outputs, wrong_preds = model.eval_model("finfile.txt")
print(result)
```
